```python
import requests
import json
import pandas as pd
import re
import numpy as np
```

```python
# fazendo requisição de API com requests
x = requests.get('https://sapl.al.am.leg.br/api/materia/materialegislativa/?ano=2023&autores=165&ementa__contains=&numero=&page=2&tipo=')
x.text
```

```python
# Dataframe contendo os códigos id da API de cada deputado
deputados = pd.read_excel('/content/drive/MyDrive/Política/dados/info_deputadosEstaduais_2023.xlsx')
deputados
```

```python
# código completo para raspar os dados das materias legislativas combinando as duas bibliotecas

# listas vazias para adição de dados requisitados da API
id = []
text = []
ano = []
ementa = []
status = []
tipo = []
nome_dep = []

# Percorrendo os indices do dataframe para requesição da API principal de materia legislativa de cada deputado
for b in deputados.index:
  # requisitando dados da API principal para pegar o número total de páginas
  x = requests.get('https://sapl.al.am.leg.br/api/materia/materialegislativa/?ano=2023&autores='+str(deputados['id_api_aleam'][b])+'&ementa__contains=&numero=&page=2&tipo=')
  json_x = json.loads(x.text)
  x_total = json_x['pagination']['total_pages']
  # printando o nome do deputado que se está fazendo a requisição da API
  print(deputados['nome'][b])

  # percorrendo o total de paginas de cada deputado
  for p in range(1,x_total):
    y = requests.get('https://sapl.al.am.leg.br/api/materia/materialegislativa/?ano=2023&autores='+str(deputados['id_api_aleam'][b])+'&ementa__contains=&numero=&page='+str(p)+'&tipo=')
    json_y = json.loads(y.text)

    # percorrendo o total de dicionarios, por pagina de cada deputado
    for a in range(0,len(json_y['results'])):

       # printando o nome do deputado
      print(deputados['nome'][b])
      nome_dep.append(deputados['nome'][b])
      # adicionando os dados vindos de results do dicionario vindos da API principal
      print(json_y['results'][a]['__str__'])
      id_json = json_y['results'][a]['id']
      id.append(id_json)
      text.append(json_y['results'][a]['__str__'])
      ano.append(json_y['results'][a]['data_apresentacao'])
      ementa.append(json_y['results'][a]['ementa'])

      # requisitando dados de status da materia legislativa da API secundaria
      sta = requests.get('https://sapl.al.am.leg.br/api/materia/materialegislativa/'+str(id_json)+'/ultima_tramitacao/')
      json_sta = json.loads(sta.text)
      print(id_json)

      try:
        print(json_sta['__str__'].split('|')[1])
        status.append(json_sta['__str__'].split('|')[1])
      except KeyError:
        status.append('Nan')
```

```python
dic = {'id':id, 'nome':nome_dep, 'texto':text, 'ano':ano, 'ementa':ementa, 'status':status}
dic
```

```python
df_dic = pd.DataFrame.from_dict(dic)
df_dic
```

```python
# salvando os dados em excel
df_dic.to_excel('materia_legislativa_deputados.xlsx')
```
